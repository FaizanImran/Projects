```python
df = pd.read_html(url, header = 0)
df
```

```python
len(df)
```

### DataFrame

```python
df = df[0]
df
```

```python
df1 = pd.read_html(url1)
df2 = pd.read_html(url2)
df3 = pd.read_html(url3)
```

```python
df1 = df1[0]
df2 = df2[0]
df3 = df3[0]
```

### Data Concat

```python
all_data = [df,df1,df2,df3]
nfl_players = pd.concat(all_data)
nfl_players
```

```python
nfl_players['Player'].duplicated().sum()
```

```python
len(nfl_players)
```

```python
nfl_players.index = np.arange(1,len(nfl_players)+1)
nfl_players
```

### Saving Data to File

```python
nfl_players.to_csv("nfl_players.csv")
```
